```python
import pandas as pd

dataset = pd.read_csv('Salary Prediction of Data Professions.csv')

dataset['DOJ'] = pd.to_datetime(dataset['DOJ'], errors='coerce')
dataset['DOJ'] = dataset['DOJ'].dt.strftime('%d-%m-%Y')


nuevos_nombres = {
    'FIRST NAME': 'Nombre',
    'LAST NAME': 'Apellido',
    'SEX': 'Genero',
    'DOJ': 'Fecha_ingreso',
    'CURRENT DATE': 'Fecha_actual',
    'DESIGNATION': 'Cargo',
    'AGE': 'Edad',
    'SALARY': 'Salario',
    'UNIT': 'Unidad',
    'LEAVES USED': 'Días_vacaciones_usados',
    'LEAVES REMAINING': 'Días_vacaciones_restantes',
    'RATINGS': 'Calificaciones',
    'PAST EXP': 'Experiencia_previa'
}

dataset.rename(columns=nuevos_nombres, inplace=True)

columnas_a_eliminar = ['Nombre','Apellido','Edad','Unidad','Genero','Fecha_ingreso','Fecha_actual'
                    ,'Días_vacaciones_usados', 'Días_vacaciones_restantes','Calificaciones']
dataset.drop(columns=columnas_a_eliminar, inplace=True)


dataset2 = pd.read_csv('Dataset salary 2024.csv')

nuevos_nombres = {
    'work_year': 'Año_trabajo',
    'experience_level': 'Nivel_experiencia',
    'employment_type': 'Tipo_empleo',
    'job_title': 'Cargo',
    'salary': 'Salario_Pais',
    'salary_currency': 'Moneda_salario',
    'salary_in_usd': 'Salario',
    'employee_residence': 'Residencia_empleado',
    'remote_ratio': 'Ratio_remoto',
    'company_location': 'Ubicación_empresa',
    'company_size': 'Tamaño_empresa'
}


dataset2.rename(columns=nuevos_nombres, inplace=True)

columnas_a_eliminar = ['Año_trabajo','Tipo_empleo','Salario_Pais','Moneda_salario','Residencia_empleado','Ratio_remoto','Tamaño_empresa']
dataset2.drop(columns=columnas_a_eliminar, inplace=True)

df_concatenado = pd.concat([dataset, dataset2], ignore_index=True)


def generalizar_cargo(cargo):
    if 'Analyst' in cargo:
        return 'Analyst'
    elif 'Engineer' in cargo:
        return 'Engineer'
    elif 'Scientist' in cargo:
        return 'Scientist'
    elif 'Business Intelligence' in cargo:
        return 'Business Intelligence'
    elif 'Data Management' in cargo:
        return 'Data Management'
    else:
        return 'Other'

def asignar_nivel_experiencia(nivel):
    if nivel <= 4:
        return 'EN'  # Entry-level
    elif nivel <= 9:
        return 'MI'  # Mid-level
    elif nivel <= 14:
        return 'SE'  # Senior
    else:
        return 'EX'  # Expert

# Aplicar la función a la columna 'Experiencia_previa'
df_concatenado['Nivel_experiencia'] = df_concatenado['Experiencia_previa'].apply(asignar_nivel_experiencia)


df_concatenado['Cargos_generalizados'] = df_concatenado['Cargo'].apply(generalizar_cargo)

#print(df_concatenado.columns)

df_concatenado.to_csv('DatosSalariosConsolidados.csv', index=False)

```
